## Test for a Difference in Two Means

------------------------------------------------------------------------

**State your statistical question in words here.**

- $H_0$: $μ_m - \mu_f = 0$: The average net wealth of male billionaires is equal to the average net wealth of female billionaires.

- $H_a$: $μ_m - \mu_f > 0 $: The average net wealth of male billionaires is greater than the average net wealth of female billionaires.

## Permutation Test

---



```r
billionaires <- read.csv(file="https://raw.githubusercontent.com/CU-Denver-MathStats-OER/Statistical-Theory/main/Data/Project-23-Fall/Billionaires-Statistics-Dataset.csv")
```

```r
female_billionaires <- billionaires[billionaires$gender == "F", "finalWorth"]
male_billionaires <- billionaires[billionaires$gender == "M", "finalWorth"]
all_billionaires <- billionaires$finalWorth
observed_diff <- mean(female_billionaires) - mean(male_billionaires)

num_permutations <- 10000
permuted_diffs <- numeric(num_permutations)

for (i in 1:num_permutations) {
  permuted_gender <- sample(billionaires$gender)
  permuted_female <- all_billionaires[permuted_gender == "F"]
  permuted_male <- all_billionaires[permuted_gender == "M"]
  permuted_diffs[i] <- mean(permuted_female) - mean(permuted_male)
}

p_value <- mean(permuted_diffs >= observed_diff)

cat("Observed Difference:", observed_diff, "\n")
cat("Permutation Test Results:\n")
cat("p-value:", p_value, "\n")
ggplot(data.frame(permuted_diffs), aes(x = permuted_diffs)) +
  geom_histogram(fill = "lightblue", color = "black", alpha = 0.7) +
  geom_vline(xintercept = observed_diff, color = "red", linetype = "dashed", size = 1) +
  labs(title = "Permutation Test Results",
       x = "Difference in Mean Net Worth (Female - Male)",
       y = "Frequency")
```

This is a permutation test of the difference in two means. The p-value is 0.5172, which is quite a large value. This means the test is likely NOT statistically significant.

## Parametric Test

---


```r
cat("Female Billionaires:\n")
print(length(female_billionaires))
print(mean(female_billionaires))

cat("\nMale Billionaires:\n")
print(length(male_billionaires))
print(mean(male_billionaires))

t_test_gender <- t.test(female_billionaires, male_billionaires, alternative = "greater")
t_test_gender

library(ggplot2)

df <- data.frame(
  Gender = rep(c("Female", "Male"), c(length(female_billionaires), length(male_billionaires))),
  NetWorth = c(female_billionaires, male_billionaires)
)

ggplot(df, aes(x = Gender, y = NetWorth, fill = Gender)) +
  geom_boxplot() +
  labs(title = "Billionaires' Net Worth by Gender",
       x = "Gender",
       y = "Net Worth",
       fill = "Gender") +
  theme_minimal()
```

```r
test_stat <- abs(t_test_gender$statistic)
deg_free <- t_test_gender$parameter

end_t <- qt(0.9997, deg_free)
xmax <- max(end_t, test_stat)
xmin <- -xmax

ggplot(NULL, aes(c(xmin, xmax))) +
  geom_area(stat = "function", fun = dt,
            args = list(df = deg_free),
            color = "black", fill = NA,
            xlim = c(-test_stat, test_stat)) +
  geom_area(stat = "function", fun = dt,
            args = list(df = deg_free),
            color = "black", fill = "firebrick2",
            xlim = c(xmin, -test_stat)) +
  geom_area(stat = "function", fun = dt,
            args = list(df = deg_free),
            color = "black", fill = "firebrick2",
            xlim = c(test_stat, xmax)) +
  geom_vline(xintercept = c(-test_stat, test_stat), linetype = "dashed",
             color = "firebrick2", linewidth = 1) +
  labs(x = "test statistic", y = "") +
  scale_y_continuous(breaks = NULL) +
  scale_x_continuous(breaks = c(-test_stat, test_stat)) +
  geom_hline(yintercept = 0) +
  theme_bw() +
  theme(axis.text.x = element_text(size = 15, color = "firebrick2"))
```

# Test for a Difference in Two Proportions

------------------------------------------------------------------------

We wish to test if there is a difference in the difference in the proportion of Billionaires who earned their fortune as entrepreneurs or founding a company and Billionaires who inherited their wealth throughout history. We will use a two sided test and a signifigance level of

$$ \alpha = 0.05. $$

- $H_0$: $ p_I - p_S = 0 $: There is no difference in the proportion of Billionaires who earned their fortune as entrepreneurs or founding a company and Billionaires who inherited their wealth throughout history.

- $H_a$: $ p_I - p_S \ne 0 $: The proportion of Billionaires who earned their fortune as entrepreneurs or founding a company is not equal to the proportion of Billionaires who inherited their wealth throughout history.




## Permutation Test

---



For our analysis of a difference in proportions, we wished to test if there was a difference in the proportion Billionaires who earned their fortune as entrepreneurs or founding a company and those who inherited their wealth. Inside the billionaire statistic dataset are the variables, selfMade and status (both stored as chr), that contain qualitative data on whether a billionaire was self-made or not. The selfMade variable only offers a binary True/False on whether a billionaire is self-made. And the publisher of the billionaires statistics dataset does not explain how they decided if a billionaire was self-made or not. The status variable, however, gives us a little more information. Once we convert status to a factor, we can see that there are 6 levels ("D", "E", "N", "R", "Split Family Fortune", and "U"). The only info offered by the publisher on these factors is "D" represents self-made billionaires (defined as Founders and Entrepreneurs) and "U" represents billionaires who inherited their wealth. For our analysis, we are using the status variable and define a self-made billionaire as an entrepreneur or founder of a company. In order to perform the permutation test for a difference in the proportion of self-made and inherited wealth billionaires, we needed to convert the status variable from a character into factor. We then had to remove four of the six levels in the status variable leaving only the "D" and "U" levels. We then finally converted the status varibale from a factor into a numeric binary variable using ifelse() allowing us to inplement our permutation test. We see this carried out in the R code cells Below.

---

We see that the proportion of billionaires who earned their fortune as entrepreneurs or founding a company is

$$ \hat{p}_e = 0.5885. $$

The proportion of billionaires who inherited their fortune is

$$ \hat{p}_i = 0.4115. $$

Thus, the observed difference in proportions, our test statistic, is

\begin{align}
  \hat{p}_{\mbox{diff}} &= \hat{p}_e - \hat{p}_i \\
  &= 0.1771.
\end{align}

If our Null Hypothesis is true, then the permutation distribution plotted below represents the Null distribution. Using our permutation distribution, we calculate our $P$-value using a two-sided test as

$$ P = 0.00002. $$

Therefore, if there is no difference in the proportion of Billionaires who earned their fortune as entrepreneurs or founding a company and Billionaires who inherited their wealth, then there is a $0.002\%$ chance of getting our observed difference in proportions. This $P$ value is much less than our significance level of $0.05$. We thus have evidence to reject the Null Hypothesis based on our permutation test. We see the permutation test and all calculations involved below.


```r
summary(billionaires.orig$status)
```

```r
status1 <- factor(billionaires.orig$status) # Convert how status is stored from chr to factor
status2 <- na.omit(droplevels(status1, c("E", "N", "R", "Split Family Fortune"))) # Removes all Levels besides "D" and "U" & omit resulting NA
status3 <- ifelse(status2 == "D", 1, 0) # Converting from a factor into a numeric binary variable
```

```r
str(status1) # Checking our work from the previous cell
str(status2)
str(status3)
```

```r
n.b <- length(status3) # Get the total number of observations in the status3 variable
n.Entrep <- sum(billionaires.orig$status == "D") # Get the total number of Entrepreneur/Founder billionaires
n.inherited <- sum(billionaires.orig$status == "U") # Get the total number of inherited wealth billionaires

n.b
n.Entrep
n.inherited
```

```r
prop.Entrep <- n.Entrep / n.b # Calculate proportion of self made billionaires
prop.inherited <- n.inherited / n.b # Calculate the proportion of inherited wealth billionaires

prop.Entrep
prop.inherited
```

```r
obsDiffProp <- prop.Entrep - prop.inherited # Observed difference in proportions
obsDiffProp
```

```r
N <- 10^5 - 1
perm.prop <- numeric(N)

for (i in 1:N)
{
  index <- sample(n.b, size = n.Entrep, replace = FALSE)
  perm.prop[i] <- mean(status3[index]) - mean(status3[-index])
}
```

```r
hist(perm.prop,
     xlab = "p_hat1 - p_hat2",
     main = "Permutation Distribution",
     col = "chartreuse3",
     xlim = c(-0.18,0.18),
     cex.lab=1.5, cex.axis=1.5, cex.main=1.5 )
abline(v = obsDiffProp, col = "red")
```

```r
p.prop <- 2*(sum(perm.prop >= obsDiffProp) + 1 ) / (N+1)
p.prop
```

## Parametric Test

---


Our sample pooled proportion is given by

\begin{align}
  \hat{p}_p &= \frac{X_e + x_i}{n.e + n.i} \\
  &= 0.5.
\end{align}

Thus, the estimate of the standard error of the Null distribution is given by

\begin{align}
  \mbox{SE}(\hat{p}_e - \hat{p}_i) &= \sqrt{ (0.5)(1-0.5) \left( \frac{1}{1223} + \frac{1}{855} \right)} \\
  &≈ 0.0223.
\end{align}

Now, we find the $z$-score as

\begin{align}
  z = \frac{ (\hat{p}_e - \hat{p}_i) - 0}{0.0223} \approx 7.9452.
\end{align}

If our Null Hypothesis is true, then the  $t$-distribution plotted below represents the Null distribution. Using our permutation distribution, we calculate our $P$-value using a two-sided test as

$$ p = 1.9385 \times 10^{-15}. $$

Therefore, if there is no difference in the proportion of Billionaires who earned their fortune as entrepreneurs or founding a company and Billionaires who inherited their wealth, then there is a $1.9385 \times 10^{-13}\%$ chance of getting our observed difference in proportions. This $P$ value is much less than our significance level of $0.05$. We thus have evidence to reject the Null Hypothesis based on our parametric test. We see the parametric test and all calculations involved below.

```r
prop.pooled <- (n.selfMade + n.inherited) / (2 * n.b)
se.b <- sqrt( prop.pooled * (1 - prop.pooled) * (1 / n.selfMade + 1 / n.inherited))

prop.pooled
se.b
```

```r
z.b <- (obsDiffProp - 0 ) / se.b
z.b
```

```r
2 * pnorm(-1*abs(z.b), 0 , 1)
```

```r
library(ggplot2)
```

```r
test.stat <- z.b  # t-test statistic
deg.free <- n.b - 1  # degrees of freedom
```

```r
################################################
# for a two-tailed test run the code cell below
################################################
v <- deg.free
end.t <- qt(0.9997, v)
test.stat <- abs(test.stat)
xmax <- max(end.t, test.stat)
xmin <- -xmax


ggplot(NULL, aes(c(xmin, xmax))) +
  geom_area(stat = "function",   fun = dt,
            args = list(df = v),
            color = "black", fill = NA,
            xlim = c(-test.stat, test.stat)) +
  geom_area(stat = "function",   fun = dt,
            args = list(df = v),
            color = "black", fill = "firebrick2",
            xlim = c(xmin, -test.stat)) +
  geom_area(stat = "function",   fun = dt,
            args = list(df = v),
            color = "black", fill = "firebrick2",
            xlim = c(test.stat, xmax)) +
  geom_vline(xintercept = c(-test.stat, test.stat), linetype="dashed",
                color = "firebrick2", linewidth = 1) +
  labs(x = "test statistic", y = "") +
  scale_y_continuous(breaks = NULL) +
  scale_x_continuous(breaks=c(-test.stat,  test.stat)) +
  geom_hline(yintercept=0) +
  theme_bw() +
  theme(axis.text.x=element_text(size=15, color = "firebrick2"))
```
